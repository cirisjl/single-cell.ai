```python
from pymongo import MongoClient
mongo_url = "mongodb://mongodb:65530"
# Connect to MongoDB using the URL
client = MongoClient(mongo_url)
# Access your database and collection
db = client["oscb"]
```

```python
jobs_collection = db.get_collection("jobs")
```

```python
jobs_collection.update_many({}, 
                            {"$rename": {
                                "process": "Process", "method": "Method", "category": "Category", "status": "Status", "created_on": "Created on", "completed_on": "Completed on"
                            }
                            })
```

```python
jobs_collection.update_many({}, 
                            {"$rename": {
                                "description": "Description"
                            }
                            })
```

```python
datasets_collection = db.datasets

unique_cell_labels = datasets_collection.distinct('Selected Cell Types.label')
unique_cell_labels
```

```python
import json

# 'w' mode opens the file for writing
with open('unique_cell_labels.json', 'w', encoding='utf-8') as f:
    json.dump(unique_cell_labels, f, indent=4)
```
